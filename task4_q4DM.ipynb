# **social networks**

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
soical_network=nx.Graph()
```

```python
soical_network.add_nodes_from([
 ('0',{'ID':'0','Name':'Nour','age':21,'gender':'Female'}),
 ('1',{'ID':'1','Name':'Alex','age':25,'gender':'Male'}),
 ('2',{'ID':'2','Name':'Ross','age':31,'gender':'Male'}),
 ('3',{'ID':'3','Name':'Rachel','age':27,'gender':'Female'}),
 ('4',{'ID':'4','Name':'Tiffany','age':22,'gender':'Female'}),
 ('5',{'ID':'5', 'Name': 'Max', 'age':30,'gender':'Male'})
 ])

soical_network.add_edges_from([
 ('0','1',{'type':'Friend'}),
 ('1','2',{'type':'Colleague'}),
 ('2','3',{'type':'Colleague'}),
 ('1','3',{'type':'Friend'}),
 ('4','3',{'type':'Friend'}),
 ('5','2',{'type': 'Friend'})
])
```

```python
# Draw the social network graph
pos = nx.spring_layout(soical_network) # Positions of the nodes
node_labels = nx.get_node_attributes(soical_network, 'Name') # Node labels
edge_labels = nx.get_edge_attributes(soical_network, 'type') # Edge labels
nx.draw_networkx_nodes(soical_network, pos, node_color='lightblue', node_size=600)
nx.draw_networkx_edges(soical_network, pos)
nx.draw_networkx_labels(soical_network, pos, labels=node_labels)
nx.draw_networkx_edge_labels(soical_network, pos, edge_labels=edge_labels)
```

```python
page_rank = nx.pagerank(soical_network)
print(page_rank)
```

```python
import community
communities = nx.algorithms.community.greedy_modularity_communities(soical_network)
print("Communities:", communities)
```

```python
# Compute degree centrality
degree_centrality = nx.degree_centrality(soical_network)
print("Degree Centrality:", degree_centrality)


# Compute betweenness centrality
betweenness_centrality = nx.betweenness_centrality(soical_network)
print("Betweenness Centrality:", betweenness_centrality)
```
