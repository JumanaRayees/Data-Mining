# **Apriori-based approach**

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Step 1: Graph Representation
# Create an empty graph
graph = nx.Graph()

# Add nodes to the graph
graph.add_nodes_from(['20', '40', '50', '60', '70'])


# Add edges to the graph
graph.add_edges_from([('20', '40'), ('20', '50'), ('40', '50'), ('40', '60'), ('50', '60'), ('60','70'),('40','70'),('20','70')])
nx.draw(graph, with_labels=True)
plt.show()
```

```python
# Step 2: Subgraph Definition
# Define the subgraph you want to discover
subgraph = nx.Graph()
subgraph.add_nodes_from(['20', '40', '50'])
subgraph.add_edges_from([('20','40'),('20','50'),('40','50')])
nx.draw(subgraph, with_labels=True)
plt.show()
```

```python
# Step 3: Support Calculation
# Count the support of the subgraph in the graph
support = 0  

for sub_nodes in nx.enumerate_all_cliques(graph):
    if subgraph.nodes() <= set(sub_nodes):
       sub = graph.subgraph(sub_nodes)
       if nx.is_isomorphic(sub, subgraph):
          support += 10
```

```python
# Step 4: Pruning
# Check if the support exceeds the threshold
support_threshold = 10 # Set a minimum support threshold
print(support)
if support >= support_threshold:
 print("Subgraph is frequent!")
else:
 print("Subgraph is not frequent.")
```
